### Add labels to data

```python
in_hospital_death_1 = pd.Series([0] * len(data_1), name="in_hospital_death")
data_1 = pd.concat([data_1, in_hospital_death_1], axis=1) 
data_1 = data_1.drop(columns=["Patient ID",  "Time [s]"])

in_hospital_death_2 = pd.Series([1] * len(data_2), name="in_hospital_death")
data_2 = pd.concat([data_2, in_hospital_death_2], axis=1) 
data_2 = data_2.drop(columns=["Patient ID",  "Time [s]"])

in_hospital_death_3 = pd.Series([0] * len(data_3), name="in_hospital_death")
data_3 = pd.concat([data_3, in_hospital_death_3], axis=1) 
data_3 = data_3.drop(columns=["Patient ID",  "Time [s]"])

in_hospital_death_4 = pd.Series([0] * len(data_4), name="in_hospital_death")
data_4 = pd.concat([data_4, in_hospital_death_4], axis=1) 
data_4 = data_4.drop(columns=["Patient ID",  "Time [s]"])

in_hospital_death_5 = pd.Series([0] * len(data_5), name="in_hospital_death")
data_5 = pd.concat([data_5, in_hospital_death_5], axis=1) 
data_5 = data_5.drop(columns=["Patient ID",  "Time [s]"])

in_hospital_death_6 = pd.Series([0] * len(data_6), name="in_hospital_death")
data_6 = pd.concat([data_6, in_hospital_death_6], axis=1) 
data_6 = data_6.drop(columns=["Patient ID",  "Time [s]"])

in_hospital_death_7 = pd.Series([1] * len(data_7), name="in_hospital_death")
data_7 = pd.concat([data_7, in_hospital_death_7], axis=1) 
data_7 = data_7.drop(columns=["Patient ID",  "Time [s]"])

data_1.shape, data_2.shape, data_3.shape, data_4.shape, data_5.shape, data_6.shape, data_7.shape, data_1.head()
```

### Merge 7 Dataframes to be 1 Dataframe

```python
merged_df = pd.concat([data_1, data_2, data_3, data_4, data_5, data_6, data_7])
print(data_1.shape, data_2.shape, data_3.shape, data_4.shape, data_5.shape, data_6.shape, data_6.shape, merged_df.shape)
```

```python
merged_df.head()
```

```python
merged_df = merged_df.set_axis([i for i in range(1, len(merged_df)+1)], axis='index')
merged_df.head()
```

### Get the data for the labels and datas

```python
labels = merged_df['in_hospital_death']
data = merged_df.drop(columns='in_hospital_death')

labels.shape, data.shape
```

```python
labels.head(), data.head()
```

### Split the datas and labels into train set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.to_numpy(), labels.to_numpy(), train_size=0.8, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Create instances of Gradient Boosting Trees and Grid Search Cross Valuation

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    "max_depth": [3,4,5,6],
    "learning_rate": [0.01, 0.001, 0.0015, 0.0001],
    'n_estimators': [700, 800, 900, 10000]
}

gbr = GradientBoostingClassifier()
grid_search = GridSearchCV(estimator=gbr, param_grid=param_grid, cv=5, n_jobs=-1)
```

### Fit the training and testing data to GridSearchCV

* When perform training locally, I set slice the data and only train with the first 100 samples. Doing this helps guarantee training time won't take indefinitely long. If you have enough computing power, please use the entire dataset for the most accuracy.

```python
grid_search.fit(X_train[:100], y_train[:100]) # when you do official training, use this much data: (X_train, y_train) -> for the most accuracy
```

### Getting the best model from GridSearchCV and the best model's parameters

```python
best_tree_model = grid_search.best_estimator_
grid_search.best_params_
```

### Evaluating the model

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc

tree_based_model_prediction = best_tree_model.predict(X_test)
# Confusion Matrix
cm = confusion_matrix(y_test, tree_based_model_prediction)
# Accuracy
accuracy = accuracy_score(y_test, tree_based_model_prediction)
# Precision
precision = precision_score(y_test, tree_based_model_prediction)
# Recall
recall = recall_score(y_test, tree_based_model_prediction)
# F1-Score
f1 = f1_score(y_test, tree_based_model_prediction)
# ROC Curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, tree_based_model_prediction)
roc_auc = auc(fpr, tpr)

print(f"Confusion matrix:\n{cm}")
print(f"Accuracy: {accuracy}")
print(f"Recall score: {recall}")
print(f"F1 score: {f1}")
```
